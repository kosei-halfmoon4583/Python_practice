各支店長は、支店の予算と実績を管理している予実管理表を本部に報告しています。
本部の担当者は、送られてきた予実管理表を集計し、上司に報告しています。
ファイルはデスクトップのMyPandasというフォルダのrawdataというところに保存されているとします。

```python
export_file_path = '/Users/watanuki/PycharmProjects/python_training/Excel_Automations/output'
import_folder_path ='/Users/watanuki/PycharmProjects/python_training/Excel_Automations/rawdata'
```

```python
path = import_folder_path + '/' + '*.xlsx'
```

```python
file_path = glob.glob(path)
```

```python
file_path
```

```python
df_concat=pd.DataFrame()

for i in file_path:
    df_read_excel=pd.read_excel(i)
    print(df_read_excel.head(3))
```

```python
df_concat=pd.DataFrame()

for i in file_path:
    df_read_excel=pd.read_excel(i)
    df_concat=pd.concat([df_read_excel, df_concat])
```

```python
df_concat
```

```python
# df_drop = df_concat.drop('Unnamed: 0', axis = 1)
```

```python
df_sort = df_concat.sort_values(by='達成率', ascending=False)
```

```python
df_sort
```

```python
df_sort.to_excel(export_file_path + '/' + '予実管理表.xlsx')
```
